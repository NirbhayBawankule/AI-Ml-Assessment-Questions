```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Load the 'tips' dataset and display the first 5 rows
tips = sns.load_dataset("tips")
print("1. First 5 rows of 'tips':\n", tips.head())
```

```python
# 2. Scatter plot of total_bill vs tip
plt.figure()
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("2. Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
# 3. Line graph for total_bill vs tip
plt.figure()
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title("3. Line Plot: Total Bill vs Tip")
plt.show()
```

```python
# 4. Bar plot: Average tip for each day
plt.figure()
sns.barplot(x='day', y='tip', data=tips, estimator='mean')
plt.title("4. Bar Plot: Avg Tip by Day")
plt.show()
```

```python
# 5. Count plot: Number of records for each day
plt.figure()
sns.countplot(x='day', data=tips)
plt.title("5. Count Plot: Records by Day")
plt.show()
```

```python
# 6. Box plot: Total bill distribution across days
plt.figure()
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("6. Box Plot: Total Bill by Day")
plt.show()
```

```python
# 7. Violin plot: Total bill vs day
plt.figure()
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("7. Violin Plot: Total Bill by Day")
plt.show()
```

```python
# 8. Histogram of total_bill with KDE
plt.figure()
sns.histplot(tips['total_bill'], kde=True)
plt.title("8. Histogram + KDE: Total Bill")
plt.show()
```

```python
# 9. Pairplot for all features in iris dataset
iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.suptitle("9. Pairplot: Iris Dataset", y=1.02)
plt.show()
```

```python
# 10. Heatmap: Numeric correlation matrix in titanic
titanic = sns.load_dataset("titanic")
plt.figure()
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("10. Heatmap: Titanic Correlation Matrix")
plt.show()
```

```python
# 11. Scatter plot of bill_length_mm vs flipper_length_mm in penguins
penguins = sns.load_dataset("penguins")
plt.figure()
sns.scatterplot(x='bill_length_mm', y='flipper_length_mm', data=penguins)
plt.title("11. Scatter Plot: Bill vs Flipper Length (Penguins)")
plt.show()
```

```python
# 12. Box plot for fare against class in titanic
plt.figure()
sns.boxplot(x='class', y='fare', data=titanic)
plt.title("12. Box Plot: Fare by Class (Titanic)")
plt.show()
```

```python
# 13. Count plot for survivors in titanic
plt.figure()
sns.countplot(x='survived', data=titanic)
plt.title("13. Count Plot: Survivors in Titanic")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# 15. Heatmap of correlation in penguins with numeric_only=True
plt.figure()
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("15. Heatmap: Penguins Correlation")
plt.show()
```

```python
# 16. Violin plot of price by cut in diamonds dataset
diamonds = sns.load_dataset("diamonds")
plt.figure()
sns.violinplot(x='cut', y='price', data=diamonds)
plt.title("16. Violin Plot: Price by Cut (Diamonds)")
plt.show()
```

```python
# 17. Pairplot of penguins dataset with hue='species'
sns.pairplot(penguins, hue='species')
plt.suptitle("17. Pairplot: Penguins by Species", y=1.02)
plt.show()
```

```python
# 19. Dark theme and scatter plot recreation
sns.set_style("dark")
plt.figure()
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("19. Dark Theme: Scatter Plot Total Bill vs Tip")
plt.show()
```

```python
# 20. Convert categorical to numeric and show full correlation heatmap
tips_numeric = tips.copy()
for col in tips_numeric.select_dtypes(include='category').columns:
    tips_numeric[col] = tips_numeric[col].astype(str)
tips_numeric_encoded = pd.get_dummies(tips_numeric, drop_first=True)
plt.figure()
sns.heatmap(tips_numeric_encoded.corr(), cmap="viridis", annot=False)
plt.title("20. Heatmap: Full Correlation (Encoded Tips)")
plt.show()
```
