```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Simple line graph
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.show()
```

```python
# 2. Add title and axis labels
plt.plot(x, y)
plt.title("Line Graph with Title and Labels")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# 3. Red dotted line
x = [0, 1, 2, 3]
y = [3, 6, 1, 8]

plt.plot(x, y, 'r--')  # Red dotted line
plt.title("Red Dotted Line")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("Two Lines in One Plot")
plt.show()
```

```python
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Two Lines with Legend")
plt.show()
```

```python
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]

plt.bar(fruits, quantity)
plt.title("Fruit Quantity Bar Chart")
plt.show()
```

```python
plt.barh(fruits, quantity)
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
subjects = ['Math', 'Science', 'English']
values = [30, 40, 30]

plt.pie(values, labels=subjects, autopct='%1.1f%%')
plt.title("Subject Distribution Pie Chart")
plt.show()
```

```python
x = [5, 7, 8, 7]
y = [8, 5, 6, 3]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
marks = [60, 70, 80, 75, 70, 90, 85, 80]

plt.hist(marks, bins=5)
plt.title("Histogram of Marks")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_title("Line Plot")

ax2.bar(['A', 'B', 'C'], [5, 7, 3])
ax2.set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Figure Size 8x5")
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 1, 5], linestyle='--', marker='o', color='green')
plt.title("Custom Style Line")
plt.show()
```

```python
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Saving as PNG")
plt.savefig("my_plot.png")
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.grid(True)
plt.title("Plot with Grid Lines")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
max_index = y.index(max(y))
plt.annotate('Highest Point', xy=(x[max_index], y[max_index]),
             xytext=(x[max_index], y[max_index]+2),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title("Annotated Max Point")
plt.show()
```

```python
bars = plt.bar(fruits, quantity)
plt.title("Bar Chart with Value Labels")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(height),
             ha='center', va='bottom')

plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [250, 300, 150, 400]

plt.bar(months, sales)
plt.xticks(rotation=45)
plt.title("Sales by Month")
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Line plot
axs[0].plot([1, 2, 3], [2, 4, 6])
axs[0].set_title("Line")

# Bar chart
axs[1].bar(['A', 'B', 'C'], [5, 3, 7])
axs[1].set_title("Bar")

# Pie chart
axs[2].pie([30, 40, 30], labels=['Math', 'Science', 'English'], autopct='%1.1f%%')
axs[2].set_title("Pie")

plt.tight_layout()
plt.show()
```

```python
plt.bar(fruits, quantity, color=['red', 'yellow', 'green'])
plt.title("Bar Chart with Custom Colors")
plt.show()
```
